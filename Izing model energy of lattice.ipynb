## Energy in Izing Model
Total energy of a 2D lattice with up and down spins

$$
E=-J \sum_{(i, j)} \sigma_{i} \sigma_{j}-H \sum_{1}^{N} \sigma_{i}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

#Creating a uniformly random array
Array=np.random.randint(0,2,(100,100))
Array
x=0
while x<Array.shape[0]:
    y=0
    while y<Array.shape[1]:
        if Array[x,y]==0:
            Array[x,y]=-1
        y=y+1
    x=x+1
print(Array)
```

```python
J=H=1 #J and H the interaction parameter values

T1=0
i=0
while i<Array.shape[0]:
    j=0
    if i==0:
        while j<(Array.shape[1]-1):
            T1=T1+(J*(Array[i,j])*(Array[i,j+1]))
            j=j+1
    if i!=0:
        while j<(Array.shape[1]-1):
            T1=T1+(J*(Array[i,j])*(Array[i,j+1]))
            j=j+1
        j=0
        while j<(Array.shape[1]):
            T1=T1+(J*(Array[i,j])*(Array[i-1,j]))
            j=j+1
    i=i+1
T1
```

```python
T2=0
x=0
y=0
while x<Array.shape[0]:
    y=0
    while y<Array.shape[1]:
        T2=T2+(H*Array[x,y])
        y=y+1
    x=x+1
T2
```

```python
print('Total Energy =',T1+T2)
```
